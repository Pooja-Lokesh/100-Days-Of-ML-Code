```python
new_train_X = train_X.reshape(train_X.shape[0], (train_X.shape[1]*train_X.shape[2]))
new_test_X = test_X.reshape(test_X.shape[0], (test_X.shape[1]*test_X.shape[2]))

print(new_train_X.shape)
print(new_test_X.shape)
```

```python
iterations = 2000
alpha = 0.01

theta_0 = np.zeros(new_train_X.shape[1])
theta_1 = np.zeros(new_train_X.shape[1])
theta_2 = np.zeros(new_train_X.shape[1])
theta_3 = np.zeros(new_train_X.shape[1])
theta_4 = np.zeros(new_train_X.shape[1])
theta_5 = np.zeros(new_train_X.shape[1])
theta_6 = np.zeros(new_train_X.shape[1])
theta_7 = np.zeros(new_train_X.shape[1])
theta_8 = np.zeros(new_train_X.shape[1])
theta_9 = np.zeros(new_train_X.shape[1])
```

```python
def gradient_descent(X_train, Y_train, theta, alpha, iterations):
  m = 60000

  for i in range(iterations):
    h = 1 / ( 1 + np.exp(- (X_train.dot(theta))))
    loss = h - Y_train
    gradient = alpha * (np.sum(X_train.T.dot(loss)) / m)
    theta = theta - gradient 

  return theta
```

```python
def predict(X, new_weights):
  Z = 1 / ( 1 + np.exp( - ( X.dot(new_weights)) ) )
  Y = np.where( Z > 0.5, 1, 0 )
  return Y
```

```python
def accuracy(Y_test, Y_pred):
  
  correctly_classified = 0
	
	# counter	
  count = 0	
  for count in range( np.size( Y_pred ) ) :
		
    if Y_test[count] == Y_pred[count] :			
      correctly_classified = correctly_classified + 1
    
    count = count + 1
  
  print( "Accuracy on test set by our model:",(correctly_classified / count ) * 100 )
```
