```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/drive/MyDrive/#100DaysofMLcode/Data.csv")
```

```python
print(data)
```

```python
# Checking for any missing values in a column 

data.isnull().values.any()
```

```python
# Counting the number of nas for the column 

data["Age"].isnull().values.sum()
```

```python
# Counting the number of nas for the column 

data["Salary"].isnull().values.sum()
```

```python
# Filling the nans with mean 

data["Age"] = data["Age"].fillna(data["Age"].mean())
```

```python
# Filling the nans with mean 

data["Salary"] = data["Salary"].fillna(data["Salary"].mean())
```

```python
print(data)
```

```python
# X - Independent attributes 
# y - Dependent attributes

X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# creating one hot encoder object with categorical feature 0
# indicating the first column
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
# Encoding the independent attribute
# Label encoder is used since the result has only two classes

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# Splitting the dataset into training and testing sets 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
# Feature scaling 
# This method helps in representing the data on a same scale
# Preserves the ranges between the data points itself, and doesn't lose any information as well. 

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```
